```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('/Users/kaciechong/Desktop/Nexus/Created -> Visualize/countrieswmost.csv')
df = df.head(10)
df['Country'] = df['Country'].str.capitalize()
df['Country'] = df['Country'].replace({
    'United kingdom - england': 'England'
})
```

```python
plt.figure(figsize=(10,6))  # Set the figure size
df.plot(kind='bar', x='Country', y='count', legend=False, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 10 Countries With the Most Programs')
plt.xticks(rotation=45, ha='right')  # Rotate the x labels for better readability

# Display the plot
plt.tight_layout()  # Adjust layout to make sure everything fits
desktop = os.path.join(os.path.expanduser('~'), 'Desktop')

save_path = os.path.join(desktop, 'Countries.png')

# Save the figure
plt.savefig(save_path, dpi=300)
```
