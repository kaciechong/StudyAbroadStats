```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('/Users/kaciechong/Desktop/SexualOrientation.csv')
```

```python
df
```

```python
df = df.rename(columns={'Unnamed: 0': 'Identity'})
df = df.rename(columns={'Academic Year of Participation': '2020-21'})
df = df.rename(columns={'Academic Year of Participation.1': '2021-22'})
df = df.rename(columns={'Academic Year of Participation.2': '2022-23'})
df = df.rename(columns={'Academic Year of Participation.3': '2023-24'})
df = df.rename(columns={'Academic Year of Participation.4': 'Grand Total'})
```

```python
df = df.drop(0)
df
```

```python
for col in df.columns[1:]:
    df[col] = df[col].str.rstrip('%').astype(float)

# Set Identity as index
df.set_index('Identity', inplace=True)

# Transpose to have years on x-axis
df = df.transpose()

# Plot
plt.figure(figsize=(10, 6))
for identity in df.columns:
    plt.plot(df.index, df[identity], marker='o', label=identity)

plt.title('Identity Distribution Over Years')
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
desktop = os.path.join(os.path.expanduser('~'), 'Desktop')

# Define save path
save_path = os.path.join(desktop, 'identity.png')

# Save the figure
plt.savefig(save_path, dpi=300)

# Show the plot
plt.show()
```
