```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('/Users/kaciechong/Desktop/Experience.csv')
df
```

```python
df = df.rename(columns={'Unnamed: 0': 'Experience'})
df = df.rename(columns={'Textual Answer': 'Not applicable/ Do not know'})
df = df.rename(columns={'Textual Answer.1': 'Very negative'})
df = df.rename(columns={'Textual Answer.2': 'Negative'})
df = df.rename(columns={'Textual Answer.3': 'Neither negative nor positive'})
df = df.rename(columns={'Textual Answer.4': 'Positive'})
df = df.rename(columns={'Textual Answer.5': 'Very positive'})
df = df.drop(0)
df
```

```python
df.set_index('Experience', inplace=True)

df = df.replace('%', '', regex=True)  # Remove '%' sign
df = df.apply(pd.to_numeric)  # Convert to numeric (as percentage)

# Plot the stacked bar chart
ax = df.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

plt.title('Perception of Study Abroad Experience by Category', fontsize=14)
plt.xlabel('Experience Type', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xticks(rotation=0)  
plt.legend(title='Experience Rating', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()  
desktop = os.path.join(os.path.expanduser('~'), 'Desktop')

save_path = os.path.join(desktop, 'Experience.png')

# Save the figure
plt.savefig(save_path, dpi=300)
```
