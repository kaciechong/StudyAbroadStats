```python
df1 = pd.read_csv('/Users/kaciechong/Desktop/Discipline.csv')
df1
```

```python
df1 = df1.rename(columns={'Unnamed: 0': 'Discipline'})
df1 = df1.rename(columns={'Academic Yearof Participation.5': 'Grand Total'})
df1 = df1.drop('Academic Yearof Participation', axis=1)
df1 = df1.drop('Academic Yearof Participation.1', axis=1)
df1 = df1.drop('Academic Yearof Participation.2', axis=1)
df1 = df1.drop('Academic Yearof Participation.3', axis=1)
df1 = df1.drop('Academic Yearof Participation.4', axis=1)

df1 = df1.drop(0)
df1
```

```python
df_cleaned = df1[df1.index % 2 == 0]
df_cleaned = df_cleaned.drop('Unnamed: 1', axis=1)
```

```python
df_cleaned
```

```python
df_cleaned['Discipline'] = df_cleaned['Discipline'].replace({
    'Agriculture & Natural Resources': 'Agriculture',
    'Architecture & Environ Design': 'Architecture',
    'Area, Ethn, Cultur & Gender Stu': 'Ethnic, Culture, & Gender',
    'Communication, Journalism & Rel': 'Communication',
    'Family & Consumer Sci/Human Sci': 'Human Sciences',
    'Public Admin & Social Serv Prof': 'Public Admin'
})

df_cleaned['Grand Total'] = df_cleaned['Grand Total'].replace({'%': '', ',': ''}, regex=True).astype(float)
```

```python
# Plot pie chart
plt.figure(figsize=(12, 10))
df_cleaned.set_index('Discipline')['Grand Total'].plot(kind='pie', autopct='%1.1f%%', legend=False)

# Title
plt.title("Percentage of Students by Discipline")

# Show the plot
plt.ylabel('')  # To hide the y-axis label
plt.tight_layout()  
desktop = os.path.join(os.path.expanduser('~'), 'Desktop')

save_path = os.path.join(desktop, 'Discipline.png')

# Save the figure
plt.savefig(save_path, dpi=300)
```
